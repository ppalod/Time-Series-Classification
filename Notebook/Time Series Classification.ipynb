### Functions used in the homework

```python
def check_delimiter_and_create_df(path):
    sniffer = csv.Sniffer()
    sep = sniffer.sniff(open(path).read()).delimiter
    if sep==",":
        df = pd.read_csv(path,skiprows=5, header=None, sep=sep)
    else:
        df = pd.read_csv(path,skiprows=5, header=None, sep=sep)
        df = df.drop(df.columns[-1],axis=1)
    df.columns=['# Columns: time','avg_rss12','var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
    return df

def read_data(f_df,folder,start,end,label):
    for p in range(start,end):
        path = "../Data/AReM/"+folder+"/dataset"+str(p)+".csv"
        if label:
            if (folder=='bending1' or folder=='bending2'):
                f_df['Label'].append(1)
            else:
                f_df['Label'].append(0)
        df = check_delimiter_and_create_df(path)
        df = df.drop(columns=df.columns[0])
        for i in range(0,6):
            f_df.setdefault('min_'+str(i+1),[]).append(df[df.columns[i]].min())
            f_df.setdefault('max_'+str(i+1),[]).append(df[df.columns[i]].max())
            f_df.setdefault('mean_'+str(i+1),[]).append(df[df.columns[i]].mean())
            f_df.setdefault('median_'+str(i+1),[]).append(df[df.columns[i]].median())
            f_df.setdefault('std_'+str(i+1),[]).append(df[df.columns[i]].std())
            f_df.setdefault('first_quartile_'+str(i+1),[]).append(df[df.columns[i]].quantile(0.25))
            f_df.setdefault('third_quartile_'+str(i+1),[]).append(df[df.columns[i]].quantile(0.75)) 

def split_dataset(l,binary,train):
    f_df={'Label':[]}
    split_list = []
    for folder,value in folders.items():
        if train:
            start=value[0]
            end=value[1]
        else:
            start=1
            end=value[0]
        for p in range(start,end):
            path = "../Data/AReM/"+folder+"/dataset"+str(p)+".csv"
            if binary:
                if (folder=='bending1' or folder=='bending2'):
                    f_df['Label'].append(1)
                else:
                    f_df['Label'].append(0)
            else:
                if (folder=='bending1' or folder=='bending2'):
                    f_df['Label'].append('bending')
                else:
                    f_df['Label'].append(folder)
            df = check_delimiter_and_create_df(path)
            df = df.drop(columns=df.columns[0])
            length = len(df)
            split=math.floor(length/l)
            data = []
            for i in range(l):
                split_df = df[(split*i):(i+1)*split]
                split_df.reset_index(drop=True, inplace=True)
                data.append(split_df)
            instance_df=pd.concat(data,axis=1,ignore_index=True)
            for i in range(0,6*l):
#                 f_df.setdefault('min_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].min())
                f_df.setdefault('mean_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].mean())
                f_df.setdefault('max_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].max())
                f_df.setdefault('third_quartile_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].quantile(0.75))
#                 f_df.setdefault('median_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].median())
#                 f_df.setdefault('std_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].std())
#                 f_df.setdefault('first_quartile_'+str(i+1),[]).append(instance_df[instance_df.columns[i]].quantile(0.25))
            split_list.append(instance_df)
    final_df=pd.concat(split_list,axis=0,ignore_index=True)
    feature_split_df = pd.DataFrame(f_df)
    feature_split_df = feature_split_df[[col for col in feature_split_df.columns if col!='Label']+['Label']]
    return feature_split_df

def plot_ROC_AUC(fpr,tpr,area):
    label_="ROC Curve : area = {} ".format(round(area, 4))
    plt.figure(figsize=(12,8))
    plt.plot(fpr, tpr, label=label_)
    plt.plot([0,1],[0,1])
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    plt.title("ROC AUC Curve")
    plt.legend()
    
def bootstrap_confidence_interval(df,repetitions,alpha):
    for col in df:
        statistics = []
        for i in range(repetitions):
            sample = df[col].sample(n=len(df[col]),replace=True)
            stat = sample.std()
            statistics.append(stat)
        statistics.sort()
        # confidence interval
        lower = np.percentile(statistics,(alpha/2)*100)
        upper = np.percentile(statistics,(1-(alpha/2))*100)
        confidence_interval[col] = [lower, upper]
        
def plot_multiple(clf, X, y, n_classes,y_score):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    y_temp = pd.get_dummies(y, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_temp[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC AUC Curve')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.3f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

### (e) Binary Classification Using L1-penalized logistic regression

### i. Repeat 1(d)iii using L1-penalized logistic regression, i.e. instead of using p values for variable selection, use L1 regularization. Note that in this problem, you have to cross-validate for both l, the number of time series into which you break each of your instances, and lambda , the weight of L1 penalty in your logistic regression objective function (or C, the budget). Packages usually perform cross-validation for lambda automatically.

```python
results = {'l_value': [], 'best_CV_score': [], 'best_C': []}
for l in range (1,21):
    df = split_dataset(l,True,True)
    X = df.drop(df.columns[-1],axis=1)
    y = df[df.columns[-1]]
#     print(X.shape)
    model = LogisticRegressionCV(cv=5, penalty="l1", solver="liblinear",random_state=42)
    model.fit(X, y)
    
    best_C = model.C_[0]
    best_C_idx = np.where(np.array(model.Cs_) == best_C)[0][0]
    best_CV_score = np.mean(model.scores_[1][:,best_C_idx])
    
    results['l_value'].append(l)
    results['best_CV_score'].append(best_CV_score)
    results['best_C'].append(best_C)
    
    
results = pd.DataFrame(results)
results
```

#### Best (l,C) pair

```python
max_score = np.amax(results["best_CV_score"])
index_of_best_l = np.where(np.array(results["best_CV_score"]) == max_score)[0][0]
best_l_value = results["l_value"][index_of_best_l]
best_c_value = results["best_C"][index_of_best_l]
print("The best (l,C) pair is:",'('+str(best_l_value)+','+str(best_c_value)+')')
```

#### Confusion matrix, ROC, and AUC of the train dataset

```python
train_df = split_dataset(best_l_value,True,True)

X_train = train_df.drop(train_df.columns[-1],axis=1)
y_train = train_df[train_df.columns[-1]]

clf = LogisticRegression(penalty='l1',solver='liblinear',C=best_c_value,random_state=42)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_train)

cf_mtx = confusion_matrix(y_pred,y_train)
print("Confusion matrix:\n")
print(cf_mtx,'\n\n')

fpr, tpr, threshold = roc_curve(y_train, y_pred)
roc_auc = auc(fpr, tpr)
area = auc(fpr, tpr)

plot_ROC_AUC(fpr,tpr,area)

print("\n\nArea under the curve:",area,'\n')
print("\nThe accuracy score is:",accuracy_score(y_train,y_pred))
```

#### Confusion matrix, ROC, and AUC of the test dataset

```python
test_df = split_dataset(best_l_value,True,False)

X_test = test_df.drop(test_df.columns[-1],axis=1)
y_test = test_df[test_df.columns[-1]]

y_pred = clf.predict(X_test)

cf_mtx = confusion_matrix(y_pred,y_test)
print("Confusion matrix:\n")
print(cf_mtx,'\n\n')

fpr, tpr, threshold = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
area = auc(fpr, tpr)

plot_ROC_AUC(fpr,tpr,area)

print("\n\nArea under the curve:",area,'\n')
print("\nThe accuracy score is:",accuracy_score(y_test,y_pred))
```

```python
results = {'l_value': [], 'best_CV_score': [], 'best_C': []}
for l in range (1,21):
    df = split_dataset(l,False,True)
    X = df.drop(df.columns[-1],axis=1)
    y = df[df.columns[-1]]
    
    le = preprocessing.LabelEncoder()
    y = le.fit_transform(y)
    
    model = LogisticRegressionCV(penalty="l1", multi_class='multinomial', solver="saga",random_state=42)
    model.fit(X, y)
    
    best_C = model.C_[0]
    best_C_idx = np.where(np.array(model.Cs_) == best_C)[0][0]
    best_CV_score = np.mean(model.scores_[0][:,best_C_idx])
    
    results['l_value'].append(l)
    results['best_CV_score'].append(best_CV_score)
    results['best_C'].append(best_C)

results = pd.DataFrame(results)
results
```

#### Best (l,C) pair

```python
max_score = np.amax(results["best_CV_score"])
index_of_best_l = np.where(np.array(results["best_CV_score"]) == max_score)[0][0]
best_l_value = results["l_value"][index_of_best_l]
best_c_value = results["best_C"][index_of_best_l]
print("The best (l,C) pair is:",'('+str(best_l_value)+','+str(best_c_value)+')')
```

#### Confusion matrix, ROC, and AUC of the train dataset

```python
train_df = split_dataset(best_l_value,False,True)

X_train = train_df.drop(train_df.columns[-1],axis=1)
y_train = train_df[train_df.columns[-1]]

le = preprocessing.LabelEncoder()
y_train = le.fit_transform(y_train)

clf = LogisticRegression(penalty='l1',multi_class='multinomial', solver='saga',C=best_c_value,random_state=42)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_train)

cf_mtx = confusion_matrix(y_pred,y_train)
print("Confusion matrix:\n")
print(cf_mtx,'\n\n')

# ROC AUC PLOT

n_classes = 6
# OnevsRest used to fit each class against all others
clf_p = OneVsRestClassifier(LogisticRegression(C=best_c_value,penalty='l1', multi_class='multinomial',solver='saga',random_state=42))

y_score = clf_p.fit(X_train, y_train).decision_function(X_train)

plot_multiple(clf_p,X_train,y_train,n_classes,y_score)


print("\nThe accuracy score is:",accuracy_score(y_train,y_pred))
```

#### Confusion matrix, ROC, and AUC of the test dataset

```python
test_df = split_dataset(best_l_value,False,False)

X_test = test_df.drop(test_df.columns[-1],axis=1)
y_test = test_df[test_df.columns[-1]]

le = preprocessing.LabelEncoder()
y_test = le.fit_transform(y_test)

y_pred = clf.predict(X_test)

cnf_matrix = confusion_matrix(y_pred,y_test)
print("Confusion matrix:\n")
print(cnf_matrix,'\n\n')

# ROC AUC PLOT

n_classes = 6
y_score = clf_p.fit(X_train, y_train).decision_function(X_test)

plot_multiple(clf_p,X_test,y_test,n_classes,y_score)

print("\nThe accuracy score is:",accuracy_score(y_test,y_pred))
```
